```python
import pandas as pd

df= pd.read_csv("lab_consumer_data.csv")
df_new1= pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_new2= pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
# copy paste functions from previous lab

def clean_my_columns(df):
    df.columns = df.columns.str.replace(" ", "_").str.lower().str.strip()
    return df

def clean_gender(x):
    if x == "Femal":
        x = "F"
    elif x == "Female":
        x = "F"
    elif x == "female":
        x = "F"
    elif x == "Male":
        x = "M"
    elif x == "male":
        x = "M"
    return x

def clean_states(x):
    if x == "AZ":
        x = "Arizona"
    elif x == "WA":
        x = "Washington"
    elif x == "Cali":
        x = "California"
    else:
        pass
    return x


def clean_edu(x):
    if x == "Bachelors":
        x = "Bachelor"
    else:
        pass
    return x

def class_fix(x):
    if x == 'Sport Car':
        x = 'Luxury'
    elif x == 'Luxury SUV':
        x = 'Luxury'
    elif x == 'Luxury Car':
        x = 'Luxury'   
    else:
        pass
    return x


def gender_numerical(x):
    if x == "F":
        x = 1
    elif x == "M":
        x = 0
    elif x == "N/A":
        x = 2
    else:
        pass
    return x
```

```python
clean_my_columns(df_new1)
display(df_new1)
df_new1=df_new1.rename(columns={"st":"state"})
```

```python
df_new1['gender']=df_new1['gender'].apply(clean_gender)
df_new1['state']= df_new1['state'].apply(clean_states)

df_new1['education']=df_new1['education'].apply(clean_edu)
df_new1['vehicle_class'] = df_new1['vehicle_class'].apply(class_fix)
```

```python
display(df_new1)
```

```python
clean_my_columns(df_new2)

df_new2=df_new2.rename(columns={"st":"state"})
```

```python

df_new2['gender']=df_new2['gender'].apply(clean_gender)
df_new2['state']= df_new2['state'].apply(clean_states)

df_new2['education']=df_new2['education'].apply(clean_edu)
df_new2['vehicle_class'] = df_new2['vehicle_class'].apply(class_fix)
```

```python
display(df_new2)
```

```python
#I added one more column with numeric gender (if needed for further analysis)

def gender_numerical(x):
    if x == "F":
        x = 1
    elif x == "M":
        x = 0
    elif x == "N/A":
        x = 2
    else:
        pass
    return x

```

```python
df_new1["gender_num"]=df_new1['gender'].apply(gender_numerical)
df_new1.head()
```

```python
df_new2["gender_num"]=df_new2['gender'].apply(gender_numerical)
df_new2.head()
```

```python
big_df= pd.concat([df, df_new1, df_new2], axis=0)
display(big_df)
```

```python
big_df.to_csv('big_df.csv')
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df2.head()
```

```python
# I assumed that CLV is the revenue generated by the customer
# that's without pivot
sales = (df2.groupby(['sales_channel'])['customer_lifetime_value'].sum()).round(2)
sales
```

```python
# and done with pivot table. 
pivot_df = (df2.pivot_table( index='sales_channel', values=['customer_lifetime_value'], aggfunc='sum')).round(2)
pivot_df

#Conculsions:
# 'Agent' and "Branch" are the most effective sales channels which might indicate that face-to-face sales are better than calling and online service
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_df2 = df2.pivot_table(index='education', columns ='gender', values=['customer_lifetime_value'], aggfunc='mean')
(pivot_df2).round(2)


# Insigths:
# CLV is the highest among people with lower education (maybe because they start to own car earlier)
# There are small differences in CLV between gender in all education groups 
# Women with lowest eduaction have the highest CLV. Men with doctor degree have the lowest CLV
```
